```python
#nested loop

n=7
for i in range(0,n):
  for j in range(0,i+1):
    print("*")
```

```python
n=7
for i in range(0,n):
  for j in range(0,i+1):
    print("*",end='')
```

```python
n=7
for i in range(0,n):
  for j in range(0,i+1):
    print("*",end="")
  print("\r")
```

```python
n=7
for i in range(0,n):
  for j in range(0,i+1):
    print("*",end="")
  print("\n")
```

```python
n=7
for i in range(0,n):
  for j in range(0,n-1):
    print("*")
    n-=1
  print("\r")
```

```python
##loop control
##break and continue
list=["apple","cherry","mango","banana"]

```

```python
for i in list:
  if i=="cherry":
    print("it is cherry")
    break
  print(i)
```

```python
for i in list:
  if i=="cherry":
    print("it is cherry")
    pass
  print(i)
```

```python
for i in list:
  if i=="cherry":
    print("it is cherry")
    continue
  print(i)
```
